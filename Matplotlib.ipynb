```python
fig = plt.figure()
# LAARGE AXES
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b);
axes.set_xlim(0,8)
axes.set_ylim(0,8000)
axes.set_xlabel('A')
axes.set_ylabel('B')
axes.set_title('String Quadraple')

#SMALL AXES
axes2=fig.add_axes([0.2,0.2,0.4,0.4])
axes2.plot(a,b);
axes2.set_xlim(1,6)
axes2.set_ylim(0,500)
axes2.set_xlabel('a')
axes2.set_ylabel('b')
axes2.set_title('String Quadraple Child')
```

```python
fig = plt.figure(dpi=200)
axes3 = fig.add_axes([0,0,1,1])
axes3.plot(a,b)
```

```python
fig = plt.figure(figsize = (2,2),dpi=200)
axes3 = fig.add_axes([0,0,1,1])
axes3.plot(a,b)
fig.savefig('new_figure.jpeg',bbox_inches='tight')
```
